```python
def unpack(sirg_author_list):
    aliases = {}
    all_velocities = {}
    all_aliases = {}
    velocities = {}
    for name in sirg_author_list:
        with open(str(name)+'_alias_dict.p','rb') as f:
            [init_alias_dict,velocity] = pickle.load(f)
        aliases[name] = init_alias_dict
        velocities[name] = velocity

    for v in aliases.values():
        all_aliases.update(v)
    for v in velocities.values():
        all_velocities.update(v)
    
    return all_aliases,all_velocities
all_aliases,all_velocities = unpack(sirg_author_list)
all_aliases
rev = {v:k for k,v in all_aliases.items()}
for k,v in all_aliases.items():
    #print(len(v))
    print(v)
```
